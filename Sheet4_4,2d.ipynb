```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

```

```python
# Countor Plot for A(q).
x= np.linspace(-5,5,50)
y= np.linspace(-5,5,50)
x,y= np.meshgrid(x,y)
z= np.sin(x**2+y**2)
plt.contour(x,y,z,cmap= 'jet')
plt.title("Contour plots- A(q)")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

```python
#Changing the resolution.
x1= np.linspace(-5,5,500)
y1= np.linspace(-5,5,500)
x1,y1= np.meshgrid(x1,y1)
z1= np.sin(x1**2+y1**2)
plt.contour(x1,y1,z1,cmap= 'jet')
plt.title("Contour plots- A(q)")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

```python
# 1. Gradient points using inbuilt quiver.
skip= (slice(None, None,3), slice(None, None,3))
plt.quiver(x[skip],y[skip],z[skip], cmap='jet')
plt.title("Direction of Gradient Pointers- A(q)")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

Here, The gradient pointers are all pointing radially outwards

```python
# 2. Gradient points using inbuilt np.gradient function.
dx, dy= np.gradient(z)
skip= (slice(None, None,3), slice(None, None,3))
plt.quiver(x[skip],y[skip], dx[skip], dy[skip], cmap='jet')
plt.title("Direction of Gradient Pointers- A(q)")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

From the plot, Clearly, the arrows are pointing in both directions, towards radially inwards and outwards

```python
# contour plot of B(q).
h= np.sin(4*x+3*y)
# Here k=px+qy= 4x+3y
plt.contour(x,y,h,cmap= 'jet')
plt.title("Contour plots- B(q)")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

```python
# Gradient pointers of B(q).
skip= (slice(None, None,3), slice(None, None,3))
plt.quiver(x,y,h, cmap='jet')
plt.title("Direction of Gradient Pointers- B(q)")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

From the figure, the arrows are pointing towards the contour lines from both directions
